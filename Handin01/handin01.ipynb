```python
import pandas as pd

df = pd.read_csv("movieData.csv")
display(df)
df.set_index("Title",inplace = True)
df.shape
df.columns
```

```python
#SPG 1
disneyMovies = df[df['Distributor'].str.contains('Disney') == True]#creater nyt datasæt der tjekker om Dis indholder Disney

#df_grossing_sorted = disneyMovies.sort_values(
#by = 'World Sales (in $)',
#ascending = False
#)#sortere ud fra world sales
top10Disney = disneyMovies.nlargest(10,'World Sales (in $)')
display(top10Disney)#henter de første 10 ud
```

```python
#SPG2
license = df['License'].unique()#gir den en unique værdi
distributers = df.iloc[:,9]#henter den det 9 index ud i collum som er Lincense

M = []
PG = []
R = []
G = []

for x in distributers:
    if x == 'G':
        G.append(x)
    if x == 'PG':
        PG.append(x)
    if x == 'R':
        R.append(x)
    if x == 'M':
        M.append(x)
    if x is None:
        pass
    
```

```python
import matplotlib.pyplot as plt
import numpy as np


diffLicense = np.array([len(G), len(PG), len(R),len(M)])
ownLabels = 'G', 'PG', 'R', 'M'

plt.pie(diffLicense, labels = ownLabels)
plt.show()
```

```python
#SPG4
genre_data = df.groupby(['Genre'])['World Sales (in $)'].mean()#avg er ultra højt tal
res = genre_data.tolist()
genres = df.iloc[:,7] # henter genre ud da den ligger på 7 col

some_dict = {} #empty dictonary arr

for key in genres:
    for value in res:
        some_dict[key] = value
        res.remove(value)
        break
```

```python
import matplotlib.pyplot as plot

 
x = list(some_dict.keys()) #x er alle keys aka genre
y = list(some_dict.values()) #y er values aka de meget høje tal

plot.bar(x, y)
plot.title("flotteste bar chart på jorden")
plot.grid()
plot.show()
```
